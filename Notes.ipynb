## Week 1 笔记

**开始日期**：04.07

**LLM 工作原理理解**：LLM 通过对大量无标注文本进行统计、预训练，将预测的句子中的下一个词作为“标签”；随后在更小规模且经过标注的目标数据集上进行微调，以遵循指令和执行分类任务。LLM 是基于 Transformer 的架构，其核心组件是注意力机制，即逐词输出时，根据需要调整输入序列的各部分的权重。
   
### **笔记**：   
1. 微调大模型最流行的两种方法是**指令**微调和**分类任务**微调。不同微调类型有其相应的标注数据集：
    - 指令微调 —— “指令-答案”对
    - 分类任务微调 —— “文本及其类别标签”
   
2. Transformer 基础架构   
   最初是为了机器翻译任务开发。   
   ![image.png](./pictures/Transformer.png)
   - 编码器：负责处理输入文本，将其编码为一系列数值表示或向量，以捕捉输入的上下文信息。
   - 解码器：接收编码向量，并据此生成输出文本。
   
   编码器和解码器都是由多层组成，这些层通过**自注意力机制**连接。
   - 自注意力机制允许模型衡量序列中不同单词或词元之间的相对重要性。
   - 它可以很好地捕捉、缓解输入数据中长距离的依赖和上下文关系。
   
3. Transformer 变体   
   ![image.png](./pictures/BERTVSGPT.png)
   -  **BERT**(Bidirectional Encoder Representations from Transformer, 双向编码预训练 Transformer)：专注于掩码预测（masked word prediction），用于情感预测、文本分类等任务，比如以前推特利用 BERT 检测有害内容。
   - **GPT**(Generative Pretrained Transformer, 生成式预训练 Transformer)：侧重于原始 Transformer 架构解码器部分，主要用于处理生成任务和生成文本序列，比如机器翻译、文本摘要、小说写作、代码编写等。最开始主要被设计和训练用于**文本补全**（text completion）任务，但它们表现出了出色的**可扩展性**。这些模型擅长执行**零样本学习任务**（在没有任何特定示例的情况下，泛化到从未见过的任务）和**少样本学习任务**（从用户提供的少量示例中进行学习）。
   ![image.png](./pictures/GPT.png)
   - GPT 本质上只包含解码器部分，并不包含编码器。由于像 GPT 这样的解码器模型是通过逐词预测
生成文本，因此它们被认为是一种**自回归模型**（autoregressive model）。自回归模型将之前的输出
作为未来预测的输入，提高了最终文本的一致性。
   
4. **注意**：并非所有的 Transformer 都是大语言模型，因为Transformer 也可用于计算机视觉领域。同样，并非所有的大语言模型都基于 Transformer 架构，因为还存在基于循环和卷积架构的大语言模型。
   
5. **词元**（Token）是模型读取文本的基本单位。

6. 这些类 GPT 模型的预训练特性使它们在针对下游任务进行微调时表现出了极高的灵活性，因此它们
也被称为**基础模型**。预训练大语言模型需要大量资源，成本极其高昂。

7. **自监督学习**（self-supervised learning）：用于下一单词预测任务，它是一种自我标记的方法，即可以使用句子或文档中的下一个词作为模型的预测标签。由于该任务允许“动态”创建标签，因此我们可以利用大量的无标注文本数据集来训练大语言模型。

8. **涌现**（emergence）：模型能够完成未经明确训练的任务的能力。

### 问题回答

1. 为什么 LLM 使用无监督预训练而非直接监督训练？   
    LLM 使用无监督预训练（更准确地说是自监督学习）的核心原因在于数据规模与获取成本的巨大差异。高质量的人工标注监督数据极其稀缺且昂贵，而互联网上存在着近乎无限的免费原始文本。通过在海量无标注文本上进行“预测下一个词”的预训练，模型能够像海绵吸水一样，低成本地学习到人类语言的语法结构、逻辑推理能力以及极其丰富的世界知识，从而构建出一个强大的“通用基座”。如果直接使用少量昂贵的监督数据从头开始训练，模型不仅难以学到通用的语言规律，还极易对特定任务产生过拟合，完全失去“举一反三”的泛化能力。
      
3. GPT 系列为什么选择解码器-only 架构？   
    GPT 系列选择纯解码器（Decoder-only）架构，是因为这种设计与自回归生成任务（即文本续写）的本质完美契合。与擅长文本理解的 Encoder（如 BERT）相比，纯 Decoder 通过因果掩码（Causal Mask）严格保证了模型只能根据前文预测后文，非常适合生成式任务；而与 Encoder-Decoder 架构相比，纯 Decoder 在工程实现和算力扩展（Scaling）上更为简洁高效。OpenAI 敏锐地发现，只要模型规模足够大，绝大部分自然语言处理任务（如翻译、问答、总结）都可以被巧妙地统一转换为“文本续写”问题，从而让这种看似简单的单向生成架构涌现出强大的通用解决能力。

      
5. RLHF 的核心思想是什么？   
    RLHF（基于人类反馈的强化学习）的核心思想是将模型的输出与人类的价值观和真实偏好进行“对齐（Alignment）”。经过预训练的模型虽然掌握了海量知识，但它的目标只是“预测最可能的下一个词”，有时会生成看似合理但实际上带有偏见、有害或答非所问的内容。RLHF 的做法是：先让人类评估员对模型的多个回答进行排序优劣打分，用这些打分训练出一个模拟人类喜好的“奖励模型（Reward Model）”；然后，使用强化学习算法（如 PPO）让 LLM 在生成文本时去追求奖励模型的最高分。这样一来，模型就从“顺着话说”转变为了“生成有用、诚实且无害的回答”。
