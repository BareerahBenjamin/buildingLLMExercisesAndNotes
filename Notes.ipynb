## Week 1 笔记

**LLM 工作原理理解**：LLM 通过对大量无标注文本进行统计、预训练，将预测的句子中的下一个词作为“标签”；随后在更小规模且经过标注的目标数据集上进行微调，以遵循指令和执行分类任务。LLM 是基于 Transformer 的架构，其核心组件是注意力机制，即逐词输出时，根据需要调整输入序列的各部分的权重。

    - 指令微调 —— “指令-答案”对
    - 分类任务微调 —— “文本及其类别标签”

8. **涌现**（emergence）：模型能够完成未经明确训练的任务的能力。

### 问题回答

1. 为什么 LLM 使用无监督预训练而非直接监督训练？   
    LLM 使用无监督预训练（更准确地说是自监督学习）的核心原因在于数据规模与获取成本的巨大差异。高质量的人工标注监督数据极其稀缺且昂贵，而互联网上存在着近乎无限的免费原始文本。通过在海量无标注文本上进行“预测下一个词”的预训练，模型能够像海绵吸水一样，低成本地学习到人类语言的语法结构、逻辑推理能力以及极其丰富的世界知识，从而构建出一个强大的“通用基座”。如果直接使用少量昂贵的监督数据从头开始训练，模型不仅难以学到通用的语言规律，还极易对特定任务产生过拟合，完全失去“举一反三”的泛化能力。
      
2. GPT 系列为什么选择解码器-only 架构？   
    GPT 系列选择纯解码器（Decoder-only）架构，是因为这种设计与自回归生成任务（即文本续写）的本质完美契合。与擅长文本理解的 Encoder（如 BERT）相比，纯 Decoder 通过因果掩码（Causal Mask）严格保证了模型只能根据前文预测后文，非常适合生成式任务；而与 Encoder-Decoder 架构相比，纯 Decoder 在工程实现和算力扩展（Scaling）上更为简洁高效。OpenAI 敏锐地发现，只要模型规模足够大，绝大部分自然语言处理任务（如翻译、问答、总结）都可以被巧妙地统一转换为“文本续写”问题，从而让这种看似简单的单向生成架构涌现出强大的通用解决能力。
        
3. RLHF 的核心思想是什么？   
    RLHF（基于人类反馈的强化学习）的核心思想是将模型的输出与人类的价值观和真实偏好进行“对齐（Alignment）”。经过预训练的模型虽然掌握了海量知识，但它的目标只是“预测最可能的下一个词”，有时会生成看似合理但实际上带有偏见、有害或答非所问的内容。RLHF 的做法是：先让人类评估员对模型的多个回答进行排序优劣打分，用这些打分训练出一个模拟人类喜好的“奖励模型（Reward Model）”；然后，使用强化学习算法（如 PPO）让 LLM 在生成文本时去追求奖励模型的最高分。这样一来，模型就从“顺着话说”转变为了“生成有用、诚实且无害的回答”。

1. 词嵌入（Text Embedding）   
   ① 嵌入：数据 → 向量，本质是将离散对象映射到连续向量空间中的点，目的是将非数值数据转换为神经网络可以处理的格式。视频、音频和文本格式需要不同的嵌入模型。
      
   >句子嵌入或段落嵌入在检索增强生成（retrieval-augmented generation）领域非常流行。通过将生成（如生成文本）与检索（如搜索外部知识库）相结合，检索增强生成能够在生成过程中引入与上下文相关的外部信息。
   
   ② **word2vec**：预训练模型，词嵌入早期最流行的方法。核心思想是出现在相似上下文中的词往往具有相似的含义；实现了根据目标词预测上下文，或根据上下文预测目标词，从而生成词嵌入。
   
   ③ 词嵌入的维度越高越能捕捉到更细微的关系，但是计算效率更高。而高维度（>3）嵌入也难以进行可视化。

   ④ LLM 模型通常会自行生成嵌入，与使用 word2vec 相比，自行生成嵌入层可以针对特定任务和数据进行优化。

    ① 为了处理词汇表之外出现的单词，会引入**特殊上下文词元**(可能包括用于标识未知词汇和文档边界的词元)来增强模型对上下文和其他相关信息的理解：

## Week 3 笔记

**开始日期**：04.22

### 笔记

1. 本周主要探讨注意力机制，然后编写自注意力机制代码。  
   
2. 注意力机制按简单到复杂分为：
    - 简化的自注意力：一种简化的自注意力技术，用于引入更广泛的概念
    - 自注意力：带有**可训练权重**的自注意力机制，这是大语言模型所使用机制的基础
    - 因果注意力：大语言模型使用的一种自注意力机制，该机制允许模型只关注序列中先前和当前的输入，从而保持文本生成的时间顺序。在自注意力的基础上增加了额外掩码，使得大语言模型可以一次生成一个单词。
    - 多头注意力：自注意力和因果注意力的一种扩展，该机制使模型能够同时关注来自不同表示子空间的信息，它将注意力机制划分成多个头，从而使模型能够并行捕获输入数据的各种特征。   
      
3. 编码器-解码器 RNN 的主要限制：在解码阶段，RNN 无法直接访问编码器中的早期隐藏状态。因此，它只能依赖当前的隐藏状态，这个状态包含了所有相关信息。这可能导致上下文丢失，特别是在复杂句子中，依赖关系可能跨越较长的距离。于是开发 Bahdanau 注意力机制（使得解码器在每个解码步骤中可以选择性地访问输入序列的不同部分）。
   
4. 在计算序列表示时，自注意力机制允许输入序列中的每个位置关注同一序列中的所有位置。自注意力机制是基于 Transformer 架构的当代大语言模型（如 GPT 系列）的关键组成部分。

5. 在自注意力机制中，“自”指的是该机制通过关联单个输入序列中的不同位置来计算注意力权重的能力。

6. 在自注意力机制中，我们的目标是为输入序列中的每个元素 x 计算上下文向量 z 。上下文向量（context vector）可以被理解为一种包含了序列中所有元素信息的嵌入向量。它们的目的是通过结合序列中其他所有元素的信息，为输入序列（如一个句子）中的每个元素创建丰富表示。

7. 实现简单自注意力机制的步骤：

   - 第一步：计算中间值 *w*，即所谓**注意力分数**。通过计算查询词元 x 与其他所有输入词元的点积来确定这些分数。
   - 第二步：对先前计算的每个注意力分数进行归一化处理，其主要目的是获得总和为 1 的**注意力权重**。这种归一化是一个惯例，有助于解释结果，并能维持大语言模型的训练稳定性。   
     在实际应用中，使用 softmax 函数进行归一化更为常见，这种方法更好地处理了极值，并在训练期间提供了更有利的梯度特性。另外，softmax 函数可以保证注意力权重总是正值，这使得输出可以被解释为概率或相对重要性，其中权重越高表示重要程度越高。
   - 第三步：通过将嵌入的输入词元 x 与相应的注意力权重相乘，再将得到的向量求和来计算**上下文向量** z 。因
此，上下文向量 z 是所有输入向量的加权总和，通过将每个输入向量与其对应的注意力权重相乘而获得。

     </div>
   
9. 缩放点积注意力的原理：对嵌入维度进行归一化是为了避免梯度过小，从而提升训练性能。例如，在类 GPT 大语
言模型中，嵌入维度通常大于 1000，这可能导致点积非常大，从而在反向传播时由于 softmax 函数的作用导致梯度非常小。当点积增大时，softmax 函数会表现得更像阶跃函数，导致梯度接近零。这些小梯度可能会显著减慢学习速度或使训练停滞。因此，通过嵌入维度的平方根进行缩放解释了为什么这种自注意力机制也被称为缩放点积注意力机制。
   
10. 因果机制：调整注意力机制，防止模型访问序列中未来的信息，每个词的预测只能依赖之前出现的词。
   
11. 多头机制：涉及将注意力机制分成多个“头”。每个头会学习数据的不同特征，使模型能够在不同的位置同时关注来自不同表示子空间的信息。

12. 因果注意力：也称为掩码注意力，是一种特殊的自注意力形式。它限制模型在处理任何给定词元时，只能基于序列中的先前和当前输入来计算注意力分数，而标准的自注意力机制可以一次性访问整个输入序列。

    <div align="center">
         <img src="./pictures/ygattention.png" width=100%>
     </div>
    在因果注意力中，获得掩码后的注意力权重矩阵的一种方法是对注意力分数应用 softmax 函数，将对角线以上的元素清零，并对所得矩阵进行归一化。

    在因果注意力中，获得掩码后的注意力权重矩阵的一种更有效的方法是在应用 softmax 函数之前将注意力分数用负无穷大值进行掩码：(softmax 函数会将其输入转换为一个概率分布。当输入中出现负无穷大值（–∞）时，softmax 函数会将这些值视为零概率。（从数学角度来看，这是因为 e–∞ 无限接近于 0。）)

    <div align="center">
         <img src="./pictures/ygattention1.png" width=100%>
     </div>
   
14. dropout 掩码：dropout 是深度学习中的一种技术，通过在训练过程中随机忽略一些隐藏层单元来有效地“丢弃”它们。这种方法有助于减少模型对特定隐藏层单元的依赖，从而避免过拟合。需要强调的是，dropout 仅在训练期间使用，训练结束后会被取消。两个特定时间点使用：一是计算注意力权重之后，二是将这些权重应用于值向量之后。

15. 多头注意力：

    ① 从因果注意力扩展到多头注意力。首先，将直观地通过堆叠多个 CausalAttention 模块来构建多头注意力模块。然后，将用一种更复杂但计算上更高效的方式来实现这个多头注意力模块。

    <div align="center">
         <img src="./pictures/dtattention.png" width=100%>
     </div>

    多头注意力的主要思想是多次（并行）运行注意力机制，每次使用学到的不同的线性投影——这些投影是通过将输入数据（比如注意力机制中的查询向量、键向量和值向量）乘以权重矩阵得到的。在代码中，可以通过实现一个简单的 MultiHeadAttentionWrapper 类来达到这一目标，MultiHeadAttentionWrapper 类堆叠了多个之前实现的 CausalAttention 模块实例。

    ② 权重划分实现多头注意力，在 MultiHeadAttention 中，计算完注意力权重和上下文向量后，将所有头的上下文向量转置为(b, num_tokens, num_heads, head_dim)的形状。这些向量接着会被重塑（展平）为(b, num_tokens, d_out)的形状，从而有效地整合所有头的输出。此外，我们在 MultiHeadAttention 中添加了一个输出投影层（self.out_proj），这是在合并多个头之后的步骤，而 CausalAttention 类中并不存在这个层。

**GPT Block 组成（用自己的话描述）**：GPT Block，也就是 Transformer 块，就像是模型处理数据的一个车间流水线，里面包含了几个核心部件：首先是用“多头注意力机制”来搞清楚一句话里各个词之间的关系，然后用带 GELU 激活函数的“前馈神经网络”对这些信息进行深加工 。在这两个主要步骤之前，会用到“层归一化”来稳住数据的分布，并在之后加上“Dropout”随机丢弃点数据来防止过拟合。最后，整个过程还会通过“快捷连接”把原始输入和加工后的输出直接加在一起，保证信号流通顺畅 。

**残差连接的作用**：残差连接，也叫快捷连接或跳跃连接。最大的作用就是给网络里的训练信号“抄近道”，专门用来对付深度神经网络里的“梯度消失”问题 。如果网络太深，误差信号在反向传播时就会一层层变弱，导致最前面的网络层根本学不到东西 。有了残差连接，它能跳过一层或几层，把前面的输入直接加到后面的输出上，相当于给梯度修了一条直达的快速通道，保证了信号能顺畅地传遍整个网络，让深层模型训练起来更稳 。

**我的 GPT-2 参数量计算**：算出来参数量和官方对不上，是因为原版 GPT-2 用了一个叫“权重共享”的省空间的方法。如果把代码里所有层的参数加起来，总共大概是 1.63 亿（163,009,536） 。但是，原版 GPT-2 为了省内存，把最开始的“词元嵌入层”和最后面的“线性输出层”共用了一套权重，因为这两个层的尺寸刚好都是 50257 乘 768 。如果按这个逻辑把输出层那部分重复的参数减掉，算出来的结果就正好是官方说的 1.24 亿（124,412,160）左右。

## Week 6 笔记

**开始日期**：05.13

### 笔记
1. 微调大语言模型的两种主要方式：用于**分类**的微调和用于**执行指令**的微调。
2. **指令微调**涉及使用特定的指令数据对一组任务进行训练，以提高语言模型理解和执行自然语言提示词中描述的任务的能力。
3. **分类微调**，即模型被训练来识别一组特定的类别标签，比如在消息中过滤“垃圾消息”和“非垃圾消息”。需要注意的是，经过分类微调的模型只能预测它在训练过程中遇到的类别。
4. 经过指令微调的模型通常能够执行更广泛的任务（多任务）。我们可以将分类微调模型视为高度专业化的模型。一般来说，开发一个专业化的模型比开发在多种任务中表现良好的通用模型更简单。
5. 对大语言模型进行分类微调的三阶段过程：第一阶段涉及准备数据集，第二阶段专注于模型设置，第三阶段涵盖模型的微调和应用。
6. 处理包含不同长度文本消息的垃圾消息数据集有以下两种方案：
   - 将所有消息截断到数据集中最短消息的长度或批次长度；
   - 将所有消息填充到数据集中最长消息的长度或批次长度。
   
   
   第一种方案计算开销更少，但如果较短的消息远小于平均长度或最长消息，那么可能会导致信息丢失，从而降低模型性能。因此，我们选择第二种方案，这样可以保留所有消息的完整内容。基于性能与效率的考虑，与其直接将字符串"<|endoftext|>"附加到每条短文本消息中，不如将与"<|endoftext|>"对应的词元 ID 添加到编码的文本消息中。
7. 分类微调过程：
   下载数据集 -> 数据集预处理 -> 创建数据加载器                                 【下载数据集】   
   -> 模型初始化 -> 加载预训练权重 -> 修改模型以进行微调 -> 实现评估工具        【模型设置】   
   -> 微调模型 -> 评估微调后的模型 -> 在新数据上使用模型                        【模型微调和应用】   
8. 由于模型已经经过了预训练，因此不需要微调所有的模型层。在基于神经网络的语言模型中，较低层通常捕捉基本的语言结构和语义，适用于广泛的任务和数据集，最后几层（靠近输出的层）更侧重于捕捉细微的语言模式和特定任务的特征。因此，只微调最后几层通常就足以将模型适应到新任务。同时，仅微调少量层在计算上也更加高效。
9. 由于分类准确率不是一个可微分的函数，这里我们使用交叉熵损失作为替代来最大化准确率。
10. 在 PyTorch 中训练深度神经网络的典型训练循环包含多个步骤，涉及对训练集中的批次进行多轮迭代。在每次循环中，我们计算每个训练集批次的损失以确定损失梯度，然后使用这些梯度来更新模型权重，以便训练集损失最小化。
11. 分类模型的评估包括计算分类准确率（正确预测的比例或百分比）。
